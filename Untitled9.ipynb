```python
import matplotlib.pyplot as plt
import numpy as np

def solve_system(a1, b1, c1, a2, b2, c2):
    """Resuelve un sistema de 2x2.
    a1*x + b1*y = c1
    a2*x + b2*y = c2
    """
    det = a1 * b2 - a2 * b1
    if det == 0:
        return None  # No hay solución única (paralelas o coincidentes)

    x = (c1 * b2 - c2 * b1) / det
    y = (a1 * c2 - a2 * c1) / det
    return (x, y)

def plot_lines(eq1_coeffs, eq2_coeffs, solution):
    """Grafica las dos líneas y el punto de solución."""
    fig, ax = plt.subplots(figsize=(8, 6))
    x = np.linspace(-10, 10, 400)

    a1, b1, c1 = eq1_coeffs
    y1 = (c1 - a1*x) / b1
    ax.plot(x, y1, label=f'{a1}x + {b1}y = {c1}')

    a2, b2, c2 = eq2_coeffs
    y2 = (c2 - a2*x) / b2
    ax.plot(x, y2, label=f'{a2}x + {b2}y = {c2}')

    if solution:
        ax.plot(solution[0], solution[1], 'ro', markersize=8, label=f'Solución: ({solution[0]:.2f}, {solution[1]:.2f})')

    ax.set_title("Gráfico de sistema de ecuaciones")
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.grid(True)
    ax.legend()
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.show()

# --- Ejemplo de uso con un ejercicio ---
# Ejercicio 1a: 5x - 2y = 1, 3x + 3y = 5
eq1 = [5, -2, 1]
eq2 = [3, 3, 5]

solution = solve_system(*eq1, *eq2)

if solution:
    print(f"Solución para el sistema: ({solution[0]:.2f}, {solution[1]:.2f})")
    plot_lines(eq1, eq2, solution)
else:
    print("El sistema no tiene una única solución.")
    plot_lines(eq1, eq2, None)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Definir el sistema de ecuaciones
# Ecuación 1: -2x + 4y = 7
# Ecuación 2: 3x - 5y = 4

# Representar el sistema en forma matricial Ax = b
# A es la matriz de coeficientes
A = np.array([[-2, 4],
              [3, -5]])

# b es el vector de términos independientes
b = np.array([7, 4])

# 2. Resolver el sistema usando NumPy
# Usamos np.linalg.solve para encontrar los valores de x e y
solucion = np.linalg.solve(A, b)
x = solucion[0]
y = solucion[1]

print(f"La solución del sistema es: x = {x:.2f} y y = {y:.2f}")

# 3. Graficar las ecuaciones
# Definimos los rangos para el eje x
x_vals = np.linspace(-10, 10, 400)

# Ecuación 1: -2x + 4y = 7  =>  4y = 2x + 7  =>  y = (2x + 7) / 4
y1_vals = (2 * x_vals + 7) / 4

# Ecuación 2: 3x - 5y = 4   =>  -5y = -3x + 4 =>  y = (3x - 4) / 5
y2_vals = (3 * x_vals - 4) / 5

# Crear el gráfico
plt.figure(figsize=(8, 6))
plt.plot(x_vals, y1_vals, label='-2x + 4y = 7')
plt.plot(x_vals, y2_vals, label='3x - 5y = 4')

# Marcar el punto de intersección
plt.plot(x, y, 'ro', label=f'Intersección ({x:.2f}, {y:.2f})')

# Personalizar el gráfico
plt.title('Solución del Sistema de Ecuaciones por Intersección Gráfica')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir el sistema a)
A = np.array([[1, 2],
              [-3, 1]])

b = np.array([1, -10])

# Resolver el sistema
try:
    solucion = np.linalg.solve(A, b)
    x, y = solucion

    print(f"La solución del sistema a) es: x = {x:.2f} y y = {y:.2f}")

    # Graficar las ecuaciones
    x_vals = np.linspace(-5, 5, 400)

    # Ecuación 1: x + 2y = 1  =>  y = (1 - x) / 2
    y1_vals = (1 - x_vals) / 2

    # Ecuación 2: -3x + y = -10 => y = 3x - 10
    y2_vals = 3 * x_vals - 10

    plt.figure(figsize=(8, 6))
    plt.plot(x_vals, y1_vals, label='x + 2y = 1')
    plt.plot(x_vals, y2_vals, label='-3x + y = -10')
    plt.plot(x, y, 'ro', label=f'Intersección ({x:.2f}, {y:.2f})')

    plt.title('Solución Gráfica del Sistema a)')
    plt.xlabel('Eje X')
    plt.ylabel('Eje Y')
    plt.grid(True)
    plt.axhline(0, color='black',linewidth=0.5)
    plt.axvline(0, color='black',linewidth=0.5)
    plt.legend()
    plt.show()

except np.linalg.LinAlgError:
    print("El sistema no tiene una solución única.")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir el sistema b)
A = np.array([[1, -4],
              [3, -12]])

b = np.array([5, 15])

# Resolver el sistema
try:
    solucion = np.linalg.solve(A, b)
    x, y = solucion

    print(f"La solución del sistema b) es: x = {x:.2f} y y = {y:.2f}")

    # Graficar las ecuaciones
    x_vals = np.linspace(-10, 10, 400)

    # Ecuación 1: x - 4y = 5 => -4y = 5 - x => y = (x - 5) / 4
    y1_vals = (x_vals - 5) / 4

    # Ecuación 2: 3x - 12y = 15 => -12y = 15 - 3x => y = (3x - 15) / 12
    y2_vals = (3 * x_vals - 15) / 12

    plt.figure(figsize=(8, 6))
    plt.plot(x_vals, y1_vals, label='x - 4y = 5')
    plt.plot(x_vals, y2_vals, label='3x - 12y = 15')

    plt.title('Solución Gráfica del Sistema b)')
    plt.xlabel('Eje X')
    plt.ylabel('Eje Y')
    plt.grid(True)
    plt.axhline(0, color='black',linewidth=0.5)
    plt.axvline(0, color='black',linewidth=0.5)
    plt.legend()
    plt.show()

except np.linalg.LinAlgError:
    print("El sistema tiene infinitas soluciones, ya que las ecuaciones son la misma línea.")

    # Graficamos para mostrar que las líneas se superponen
    x_vals = np.linspace(-10, 10, 400)

    y_vals = (x_vals - 5) / 4

    plt.figure(figsize=(8, 6))
    plt.plot(x_vals, y_vals, 'b-', label='x - 4y = 5')
    plt.plot(x_vals, y_vals, 'r--', label='3x - 12y = 15 (Superpuesta)')

    plt.title('Infinitas Soluciones (Líneas Coincidentes)')
    plt.xlabel('Eje X')
    plt.ylabel('Eje Y')
    plt.grid(True)
    plt.axhline(0, color='black',linewidth=0.5)
    plt.axvline(0, color='black',linewidth=0.5)
    plt.legend()
    plt.show()
```

c) ¿Qué relación hay entre los puntos de la recta y las soluciones de la ecuación?
La relación es directa y fundamental: cada punto en la recta es una solución de la ecuación. La recta es la representación gráfica de todas las soluciones posibles de la ecuación lineal. Cualquier par de coordenadas (x,y) que se encuentre en la recta, satisface la ecuación.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir el sistema 21
A = np.array([[-1, 1],
              [-2, 2]])

b = np.array([5, 2])

# Resolver el sistema
try:
    np.linalg.solve(A, b)
except np.linalg.LinAlgError:
    print("El sistema no tiene solución única porque las líneas son paralelas.")

# Graficar las ecuaciones
x_vals = np.linspace(-5, 5, 400)
y1_vals = x_vals + 5
y2_vals = x_vals + 1

plt.figure(figsize=(8, 6))
plt.plot(x_vals, y1_vals, label='-x + y = 5')
plt.plot(x_vals, y2_vals, label='-2x + 2y = 2')

plt.title('Solución Gráfica del Ejercicio 21 (Líneas Paralelas)')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Definir el sistema de ecuaciones en forma matricial
# Ax = b
A = np.array([[2, 1],
              [-2, 1]])

b = np.array([12, 4])

# 2. Resolver el sistema
solucion = np.linalg.solve(A, b)
a, b_val = solucion

print(f"La cifra de las centenas y unidades (a) es: {a:.0f}")
print(f"La cifra de las decenas (b) es: {b_val:.0f}")
print(f"El número capicúa es: {int(a) * 100 + int(b_val) * 10 + int(a)}")

# 3. Graficar las ecuaciones
# Despejar 'b' en cada ecuación para graficar (como si 'a' fuera 'x' y 'b' fuera 'y')
# Ecuación 1: b = 12 - 2a
# Ecuación 2: b = 4 + 2a

a_vals = np.linspace(-2, 6, 400)
b1_vals = 12 - 2 * a_vals
b2_vals = 4 + 2 * a_vals

plt.figure(figsize=(8, 6))
plt.plot(a_vals, b1_vals, label='$2a + b = 12$')
plt.plot(a_vals, b2_vals, label='$-2a + b = 4$')
plt.plot(a, b_val, 'ro', label=f'Intersección ({a:.0f}, {b_val:.0f})')

plt.title('Solución Gráfica del Problema 10')
plt.xlabel('Cifra de las centenas (a)')
plt.ylabel('Cifra de las decenas (b)')
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir el sistema de ecuaciones
# Ecuación 1: b + a = 11  ->  a + b = 11
# Ecuación 2: b - a = 5   -> -a + b = 5

A = np.array([[1, 1],
              [-1, 1]])

b = np.array([11, 5])

# Resolver el sistema
solucion = np.linalg.solve(A, b)
a_sol, b_sol = solucion

print(f"La altura (a) es: {a_sol:.0f} cm.")
print(f"La base (b) es: {b_sol:.0f} cm.")

# Graficar las ecuaciones (cambiando 'a' por 'x' y 'b' por 'y' para el gráfico)
x_vals = np.linspace(-10, 15, 400)
y1_vals = 11 - x_vals
y2_vals = x_vals + 5

plt.figure(figsize=(8, 6))
plt.plot(x_vals, y1_vals, label='Perímetro: b + a = 11')
plt.plot(x_vals, y2_vals, label='Relación: b - a = 5')
plt.plot(a_sol, b_sol, 'ro', label=f'Solución ({a_sol:.0f}, {b_sol:.0f})')
plt.title('Solución Gráfica del Problema 11')
plt.xlabel('Altura (a)')
plt.ylabel('Base (b)')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir el sistema de ecuaciones
# Ecuación 1: x + y = 40
# Ecuación 2: 0.94x + 0.86y = 35.6

A = np.array([[1, 1],
              [0.94, 0.86]])

b = np.array([40, 35.6])

# Resolver el sistema
solucion = np.linalg.solve(A, b)
x_sol, y_sol = solucion

print(f"Los litros del primer líquido son: {x_sol:.0f} L.")
print(f"Los litros del segundo líquido son: {y_sol:.0f} L.")

# Graficar las ecuaciones
x_vals = np.linspace(0, 40, 400)
y1_vals = 40 - x_vals
y2_vals = (35.6 - 0.94 * x_vals) / 0.86

plt.figure(figsize=(8, 6))
plt.plot(x_vals, y1_vals, label='Volumen total: x + y = 40')
plt.plot(x_vals, y2_vals, label='Costo total: 0.94x + 0.86y = 35.6')
plt.plot(x_sol, y_sol, 'ro', label=f'Solución ({x_sol:.0f}, {y_sol:.0f})')
plt.title('Solución Gráfica del Problema 12')
plt.xlabel('Litros del 1er líquido (x)')
plt.ylabel('Litros del 2do líquido (y)')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.show()
```
